```python
import pandas as pd
import glob
```

# Create the mega file

```python
excel_files = glob.glob('./NY/*.xlsx') #/Users/trananhngan/Downloads/vortyx/NY
```

```python
print(excel_files[0])
print(len(excel_files))
```

```python
all_df = []
for file in excel_files:
    df = pd.read_excel(file)
    all_df.append(df)

# Concaternate 
mega = pd.concat(all_df, ignore_index = True)
```

```python
mega.head()
```

```python
NY_mega = mega[['Company Name', 'Line of Business', 'Year of Founding', 'Phone No']]
NY_mega.head()
```

# Data wrangling: missing values

```python
# Drop rows where Phone No has Null values
NY_mega = NY_mega.dropna(subset = ['Phone No'])
print(NY_mega.isnull().sum())
```

# Data wrangling: duplicates

```python
# Find duplicates in the Company Name column 
# Find all duplicate values in the "Company Name" column
duplicate_values = NY_mega['Company Name'][NY_mega['Company Name'].duplicated(keep=False)]

# Filter the DataFrame to include only rows with duplicate values in the "Company Name" column
duplicate_rows = NY_mega[NY_mega['Company Name'].isin(duplicate_values)]

duplicate_rows_sorted = duplicate_rows.sort_values(by='Company Name')

display(duplicate_rows_sorted)
```

```python
NY_mega = NY_mega.drop_duplicates(subset=['Company Name'], keep = 'first') 
```

```python
NY_mega = NY_mega.drop_duplicates(subset=['Phone No'], keep = 'first') 
```
